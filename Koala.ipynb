In this blog post, we will discuss Berkeley University’s Koala, a research prototype chatbot fine-tuned using Meta’s LLaMA on dialogue data curated from the web. We detail the dataset curation and training process of our model and present the results of a user study comparing Koala to ChatGPT and Stanford’s Alpaca. Results show that Koala can effectively respond to a range of user queries, generating responses that are often preferred over Alpaca and, in some cases, tied with ChatGPT.

Datasets: Focus on a small high-quality dataset
A significant challenge in building dialogue models is curating training data. Popular chat models like ChatGPT rely on proprietary datasets that are costly to annotate. Dialogue data from the web and public datasets, including dialogues with other large language models such as ChatGPT, were collected to build Koala.

Rather than collecting as much web data (scrapping) as possible, the focus was on gathering a small high-quality dataset. Public datasets were used for question answering, human feedback, and dialogues with existing language models. Dataset includes

1. Public ChatGPT Data
– ShareGPT — Public User-Shared Dialogues with ChatGPT (30k)
– Human ChatGPT Comparison Corpus-HC3 (87k)
2. Open Source Data:
– Open Instruction Generalist-OIG (30k)
– Stanford Alpaca (52k)
– Anthropic HH(160k)
– OpenAI WebGPT (20k)
– OpenAI Summarization (93k).
